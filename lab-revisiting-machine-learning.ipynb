```python
#Identify columns that over 85% missing values

def threshold (data, perc= 0.85):
    nulls_percent_df= pd.DataFrame(data.isna().sum()/len(data)).reset_index()
    nulls_percent_df.columns = ['column_name', 'nulls_percentage']
    columns_above_threshold = nulls_percent_df[nulls_percent_df['nulls_percentage']>perc]
    drop_columns_list = list(columns_above_threshold['column_name'])
    return drop_columns_list
```

```python
 drop_columns_list= threshold(df)
```

```python
def clean_gender_column(x):
    if x in ["U", "J", "C", "A"]:
        return "others"
    else:
        return x
    
df["GENDER"] = df["GENDER"].apply(clean_gender_column) # clean_gender_column(df["GENDER"])
```
